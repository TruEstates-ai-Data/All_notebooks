```python
def write_to_excel(df, sheet_name, file_name):
    try:
        book = load_workbook(file_name)
    except (BadZipFile, FileNotFoundError): # added FileNotFoundError handling
        # If the file is corrupted or not an Excel file, or the file doesn't exist, create a new workbook
        book = openpyxl.Workbook()  # Create a new workbook
        # Save the new workbook immediately to ensure it's a valid file
        book.save(file_name) # This line was added to save the new workbook

    # Passing the workbook to ExcelWriter using engine_kwargs:
    writer = pd.ExcelWriter(file_name, engine='openpyxl', mode='a')

    # If sheet exists, remove it
    if sheet_name in book.sheetnames:
        book.remove(book[sheet_name])

    # Write the DataFrame to the specified sheet
    # Check if df is a DataFrame before calling to_excel
    if isinstance(df, pd.DataFrame):
        df.to_excel(writer, sheet_name=sheet_name, index=True)
    else:
        print(f"Error: 'df' is not a DataFrame. It is of type: {type(df)}")  # Print an error message
    writer.close()


# Assuming grouped_stats was supposed to be created like this:
#grouped_stats = df.groupby('instance_date')[['actual_worth','meter_sale_price']].describe().round(4)


# Call the function
#write_to_excel(grouped_stats, sheet_name='Year_wise')
```
