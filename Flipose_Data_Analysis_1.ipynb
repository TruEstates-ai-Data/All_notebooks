```python
!pip install --upgrade pandas openpyxl
import pandas as pd
import openpyxl
from zipfile import BadZipFile # Importing the BadZipFile exception


def relation(df, file_name='relation_output.xlsx'):
    """
    Analyzes relationships between categorical columns and 'trans_X_3',
    stores results in an Excel file with separate sheets for each column.

    Args:
        df (pd.DataFrame): The input DataFrame.
        file_name (str, optional): The name of the Excel file. Defaults to 'relation_output.xlsx'.
    """

    cat_col = df.select_dtypes(include='object')
    trans = ['transaction_id']
    cat_col = cat_col.drop(columns=trans)

    # Try to open the file, create a new one if it doesn't exist or is corrupted
    try:
        book = openpyxl.load_workbook(file_name)
    except (openpyxl.utils.exceptions.InvalidFileException, FileNotFoundError, BadZipFile): # added FileNotFoundError handling
        book = openpyxl.Workbook()
        book.save(file_name)  # Save the new workbook immediately

    # Instead of passing 'workbook' directly, use engine_kwargs
    # Pass the workbook to the ExcelWriter using engine_kwargs
    writer = pd.ExcelWriter(file_name, engine='openpyxl',  mode='a')

    # Remove sheet if it already exists
    if file_name in writer.book.sheetnames:  # Remove sheet if it already exists
        writer.book.remove(writer.book[file_name])

    for col in cat_col:
        grouped = df.groupby(['trans_X_3', col])[col].count().reset_index(name='count')
        sorted_grouped = grouped.sort_values(by=['count', col], ascending=[False, True])

        # Write to Excel sheet
        sorted_grouped.to_excel(writer, sheet_name=col, index=False)

    writer.save()
    writer.close()

# Example usage:
relation(df)
```

```python
def outliers(df):
    num_col = df.select_dtypes(include='number')
    Q1 = df['procedure_area'].quantile(0.25)
    Q3 = df['procedure_area'].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    return upper_bound, lower_bound

```
